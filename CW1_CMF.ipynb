```python
# Example of usage of NumPy
import numpy as np

# Creating a 1D array and a 2D matrix
array_1d = np.array([1, 2, 3, 4, 5])
matrix_2d = np.array([[1, 2], [3, 4], [5, 6]])

# Basic mathematical operations
array_squared = array_1d ** 2  # Element-wise square
array_sum = np.sum(array_1d)  # Sum of array elements
matrix_transpose = np.transpose(matrix_2d)  # Transposing a matrix

print("Original Array:", array_1d)
print("Squared Array:", array_squared)
print("Sum of Array:", array_sum)
print("Original Matrix:\n", matrix_2d)
print("Transposed Matrix:\n", matrix_transpose)
```

```python
# Example of usage of Pandas
import pandas as pd

# Creating a sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24, 27, 22, 32],
    'Score': [85, 91, 78, 88]
}
df = pd.DataFrame(data)

# Basic DataFrame operations
df['Age_Squared'] = df['Age'] ** 2  # Adding a new column with squared age
average_score = df['Score'].mean()  # Calculating the average score

print("DataFrame:\n", df)
print("Average Score:", average_score)
```

```python
# Example of usage of yfinance
import yfinance as yf

# Retrieving historical data for Nvidia (NVDA)
ticker = yf.Ticker("NVDA")
data = ticker.history(period = "1y")  # Last 1 year of data

# Displaying the retrieved data
print(data.head())

# Plotting the closing price over time
plt.plot(data.index, data['Close'], label = "Nvidia Closing Price")
plt.xlabel("Date")
plt.ylabel("Close Price (USD)")
plt.title("Nvidia Stock Closing Price Over the Last Month")
plt.xticks(rotation = 45) # Rotate x-axis dates for better readability
plt.legend()
plt.show()
```

```python
# Import datetime to handle date and time operations
import datetime

# Define the tickers for BAE Systems (BA.L) and GSK (GSK.L)
tickers = ['BA.L', 'GSK.L']

# Define the time period
end_date = datetime.datetime.now()
start_date = end_date - datetime.timedelta(days = 2 * 365)  # Approximately two years from now

# Download the data
data = yf.download(tickers, start = start_date, end = end_date)

# Display the first few rows of the data
print(data.head())
```

```python
# Plotting the closing price over time
plt.figure(figsize = (18, 8))
plt.plot(data['Close']['BA.L'], color = 'blue',label = 'BA.L (BP)', linewidth = 2)
plt.plot(data['Close']['GSK.L'], color = 'orange',label = 'GSK.L (GSK)', linewidth = 2)
plt.title('Stock Prices of BAE Systems and GSK over the last two years')
plt.xlabel('Date')
plt.xticks(rotation = 45)
plt.ylabel('Price (GBP)')
plt.legend()
plt.grid()
plt.show()
```

```python
# Calculate daily returns
daily_returns = data['Adj Close'].pct_change()
print(daily_returns.head())

# Remove NaN values
daily_returns = daily_returns.dropna()

# Calculate annualised average returns
# 252 represents the approximate number of trading days in a year for the LSEG
annualised_avg_returns = daily_returns.mean() * 252

# Calculate annualised standard deviation
annualised_std_dev = daily_returns.std() * np.sqrt(252)

# Combine the results into a DataFrame for display
results_df = pd.DataFrame({'Annualised Average Return': annualised_avg_returns,
                           'Annualised Standard Deviation': annualised_std_dev})

# Display the results
print(results_df)
```
